```python
spark.conf.set(
    "fs.azure.account.key.nehastrgacc.dfs.core.windows.net",
    ""
)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
raw_path="abfss://raw@nehastrgacc.dfs.core.windows.net/DataSet"

sales_df = (
    spark.read
    .option("header",True)
    .option("inferSchema",True)
    .csv(raw_path)
)

bronze_df = (
    sales_df
    .withColumn(
        "ingestion_timestamp",
        current_timestamp()
    )
)

bronze_path="abfss://bronze@nehastrgacc.dfs.core.windows.net/DataSet"

bronze_df.write \
.format("delta") \
.mode("overwrite") \
.save(bronze_path)
```
