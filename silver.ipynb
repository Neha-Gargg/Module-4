```python
spark.conf.set(
    "fs.azure.account.key.nehastrgacc.dfs.core.windows.net",
    ""
)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
bronze_path="abfss://bronze@nehastrgacc.dfs.core.windows.net/DataSet"

bronze = spark.read.format("delta").load(bronze_path)

silver_df = bronze.dropna()

silver_df = (
    silver_df
    .dropDuplicates(["OrderID"])
)

silver_df = (
    silver_df
    .withColumnRenamed(
        "CustomerName",
        "customer_name"
    )
    .withColumnRenamed(
        "OrderDate",
        "order_date"
    )
)

silver_path="abfss://silver@nehastrgacc.dfs.core.windows.net/CleanDataSet"


silver_df.write \
.format("delta") \
.mode("overwrite") \
.save(silver_path)
```
